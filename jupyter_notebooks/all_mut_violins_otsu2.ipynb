```python
import json
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

# G1 and M data, CDC14, Otsu

```python
mutants = ['wt', 'ycg1d2', 'fob1d']
mutantList = []
labels = ['CBF5', 'CDC14']
labelList = []
phases = ['G1', 'M']
phaseList = []
areaList = []
fileStrings = []
for label in labels:
    for phase in phases:
        fileStr = 'json_' + label + '_' + 'otsu2' + '_data_' + phase + '.json'
        file = open(fileStr)
        fileDict = json.load(file)
        for mutant in mutants:
            areaList.extend(fileDict[mutant])
            mutantList.extend([mutant] * len(fileDict[mutant]))
            labelList.extend([label] * len(fileDict[mutant]))
            phaseList.extend([phase] * len(fileDict[mutant]))
totalDict = {'areas': areaList, 'mutant': mutantList, 'label': labelList, 'phase': phaseList}
df = pd.DataFrame.from_dict(totalDict)
```

```python
cdc14_g1_filtered_otsu = df.loc[(df['label']=='CDC14') & (df['phase']=='G1')]
ax = sns.violinplot(x='mutant', y='areas', data=cdc14_g1_filtered_otsu, inner='quartile', cut=0, palette='Set2')
ax.set(ylabel='CDC14-GFP Area (Pixels)', xlabel='', ylim=(0, 800))
fig = ax.get_figure()
fig.set_size_inches(3, 2.5)
fig.savefig('cdc14_g1_otsu2.png', bbox_inches='tight', dpi = 2000)
```

```python
cdc14_M_filtered_otsu = df.loc[(df['label']=='CDC14') & (df['phase']=='M')]
ax = sns.violinplot(x='mutant', y='areas', data=cdc14_M_filtered_otsu, inner='quartile', cut=0, palette='Set2')
ax.set(ylabel='CDC14-GFP Area (Pixels)', xlabel='', ylim=(0, 800))
fig = ax.get_figure()
fig.set_size_inches(3, 2.5)
fig.savefig('cdc14_m_otsu2.png', bbox_inches='tight', dpi = 2000)
```
